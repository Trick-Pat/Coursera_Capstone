# Introduction/ Business Problem

Imagine you are a student living in Vancouver, BC, Canada and you want to offer some street art tours to earn some money. Because there is a lot of street art spread all around Vancouver you have to define a route for your tour. Because you want to offer a walking tour you need to determine a neighboorhood with a high density of street art and short distances between the different artworks. You should also consider the popularity of the different artworks so that you don't just cosider the total number of them. Sometimes there might be a huge number of small and unspectatcular artworks so that you won't meet the expectations of the people attending your tour. The attendees want to see the most beautiful, spectacular and popular street artworks of Vancouver,BC and they don't want the tour to take longer than 2 hours. Also they don't want to walk more than 8 miles.
